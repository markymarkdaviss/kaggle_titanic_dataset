```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt

%matplotlib inline
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
# Setting the value of Male to 1 and value of female to 0
train_df['Male'] = [0 if value == 'female' else 1 for value in train_df['Sex']]
```

```python
# setting the null values to 
train_df['Age'].fillna(train_df['Age'].median(), inplace=True)
```

```python
sns.distplot(train_df['Fare'])
```

```python
# cutting values into different bins
train_df['Faregrp'] = pd.cut(train_df['Fare'], [0, 25, 50, 80, 500], labels=[1,2,3,4])
```

```python
# figuring out percentage of survived from Fare
pd.value_counts(train_df['Faregrp'])
```
